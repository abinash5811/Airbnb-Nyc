### let's do something with missing values and make the dataset more appropriate for analyse.

```python
# As there are only 16 values are missing in the 'name' column we can drop those values.
df.dropna(subset=['name'], inplace=True)

# Verify the changes
print(df)
```

```python
# also here as the 'host_name' has only 21 values missing we can drop those values too, as they wont matter much.
df.dropna(subset=['host_name'],inplace=True)
df
df.isna().sum()
```

```python
# as the dtype of 'last_review' column is object but the data are in datetime fromatwe need to change the column to pandas datetime data type
df['last_review']=pd.to_datetime(df['last_review'])

# now fill the null values with mode dates(most frequently occuring dates)
mode_date=df['last_review'].mode()[0]
df['last_review'].fillna(mode_date,inplace=True)
```

*the mode is used to fill the missing values in the 'last_review' column instead of mean or median because the 'last_review' column contains dates, which are categorical in nature. Unlike numerical data, taking the mean or median of dates may not provide meaningful information.*

*The mode represents the most frequently occurring date in the 'last_review' column. By filling the missing values with the mode date, we are essentially replacing the missing values with the most common date found in the dataset.*

```python
#let's repalce the missing values of 'reviews_per_month' column with it's mean value.
mean_reviews_per_month=df['reviews_per_month'].mean()
df['reviews_per_month'].fillna(mean_reviews_per_month,inplace=True)
print(df['reviews_per_month'])
```

```python


# Convert the 'last_review' column to datetime
df['last_review'] = pd.to_datetime(df['last_review'])

# Extract year and quarter from the 'last_review' column
df['year'] = df['last_review'].dt.year
df['quarter'] = df['last_review'].dt.quarter

# Group by year and quarter and calculate the average availability
availability_by_quarter = df.groupby(['year', 'quarter'])['availability_365'].mean().reset_index()

# Sort the data by year and quarter
availability_by_quarter.sort_values(['year', 'quarter'], inplace=True)

# Plot the availability over time using a bar chart
plt.figure(figsize=(10, 6))
plt.bar(availability_by_quarter['year'].astype(str) + 'Q' + availability_by_quarter['quarter'].astype(str), availability_by_quarter['availability_365'])
plt.xlabel('Year and Quarter')
plt.ylabel('Average Availability (in days)')
plt.title('Availability of Listings by Quarter')
plt.xticks(rotation=45)
plt.show()


```

**Seasonal Availability**: *By observing the heights of the bars, you can identify any seasonal patterns in the availability of listings. For example, if certain quarters consistently have higher or lower average availability, it may indicate periods of high demand or low demand in the market.*

**Demand Fluctuations**: *The bar chart allows you to analyze the fluctuations in availability over time. Sharp changes in the average availability between quarters may reflect shifts in market demand, which can be influenced by factors such as holidays, events, or tourist seasons.*

**Booking Patterns**: *The availability trends can provide insights into booking patterns. If you notice consistent patterns where availability is lower in certain quarters, it might suggest higher booking activity during those periods*.

*These insights can guide hosts, property managers, and guests in making informed decisions regarding pricing, marketing, and planning their bookings.*

```python


# Calculate the correlation matrix
corr_matrix = df.corr()

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

### *Here The heatmap allows us to visualize the strength and direction of the correlation between different pairs of variables. Positive values indicate a positive correlation, negative values indicate a negative correlation, and values close to 0 indicate a weak or no correlation.*
